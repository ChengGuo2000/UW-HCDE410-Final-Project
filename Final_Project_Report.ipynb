## Data Exploration - Process and Findings

#### Preparation

```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Data
final_dataset = pd.read_csv("MushroomDataset/secondary_data.csv", sep = ";")

# Check NA values
final_dataset.isna().sum()
# The following columns contain NA value:
# cap-surface, gill-attachment, gill-spacing, stem-root, 
# stem-surface, veil-type, veil-color, ring-type, spore-print-color

# List Features to Explore
color_features = ["cap-color", "gill-color", "stem-color"]
size_features = ["cap-diameter", "stem-height", "stem-width"]
```

#### Qualitative Feature - Cap Color

```python
poisonous = list(final_dataset[final_dataset['class'] == 'p']['cap-color'])
edible = list(final_dataset[final_dataset['class'] == 'e']['cap-color'])
poisonous_dict = {x:poisonous.count(x) for x in poisonous}
edible_dict = {x:edible.count(x) for x in edible}
X = np.arange(len(poisonous_dict))
ax = plt.subplot(111)
ax.bar(X-0.2, poisonous_dict.values(), width=0.4, color='r', align='center')
ax.bar(X+0.2, edible_dict.values(), width=0.4, color='g', align='center')
ax.legend(('Poisonous','Edible'))
plt.xticks(X, poisonous_dict.keys())
plt.title("Bar Chart of Edibility and Cap Color")
plt.savefig("Generated_Graphics/Bar_Cap-Color.jpg")
plt.show()
```

## Data Modeling - Process and Findings

#### Preparation

```python
# Import Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Load Data
final_dataset = pd.read_csv("MushroomDataset/secondary_data.csv", sep = ";")

# Extract Features
color_features = ["cap-color", "gill-color", "stem-color"]
size_features = ["cap-diameter", "stem-height", "stem-width"]
other_features = ["cap-shape", "does-bruise-or-bleed", "has-ring", "habitat", "season"]
color_size_features = color_features + size_features
all_available_features = color_size_features + other_features

# Change Data Types
letter_dict = {"a":1, "b":2, "c":3, "d":4, "e":5, "f":6, "g":7,
               "h":8, "i":9, "j":10, "k":11, "l":12, "m":13, "n":14,
               "o":15, "p":16, "q":17, "r":18, "s":19, "t":20, 
               "u":21, "v":22, "w":23, "x":24, "y":25, "z":26}
def change_type(x):
    return letter_dict[x]
for feature in color_features:
    final_dataset[feature] = final_dataset.apply(lambda x: change_type(x[feature]), axis=1)
for feature in other_features:
    final_dataset[feature] = final_dataset.apply(lambda x: change_type(x[feature]), axis=1)
    
# Split Data
train_data, test_data = train_test_split(final_dataset, test_size = 0.2)
```

#### Using only color and size features - Logistic Regression

```python
log_model = LogisticRegression()
log_model.fit(train_data[color_size_features], train_data['class'])
log_pred = log_model.predict(test_data[color_size_features])
log_cm = confusion_matrix(test_data['class'], log_pred)
log_disp = ConfusionMatrixDisplay(confusion_matrix=log_cm)
log_disp.plot()
plt.title('Confusion Matrix of Logistic Regression with only Color and Size Features')
plt.savefig("Generated_Graphics/CM_LR_ColorSize.jpg")
plt.show()
```

#### Using only color and size features - K Nearest Neighbors

```python
knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(train_data[color_size_features], train_data['class'])
knn_pred = knn_model.predict(test_data[color_size_features])
knn_cm = confusion_matrix(test_data['class'], knn_pred)
knn_disp = ConfusionMatrixDisplay(confusion_matrix=knn_cm)
knn_disp.plot()
plt.title('Confusion Matrix of K Nearest Neighbors with only Color and Size Features')
plt.savefig("Generated_Graphics/CM_KNN_ColorSize.jpg")
plt.show()
```

#### Using only color and size features - AdaBoost

```python
ab_model = AdaBoostClassifier(n_estimators = 10)
ab_model.fit(train_data[color_size_features], train_data['class'])
ab_pred = ab_model.predict(test_data[color_size_features])
ab_cm = confusion_matrix(test_data['class'], ab_pred)
ab_disp = ConfusionMatrixDisplay(confusion_matrix=ab_cm)
ab_disp.plot()
plt.title('Confusion Matrix of AdaBoost with only Color and Size Features')
plt.savefig("Generated_Graphics/CM_AB_ColorSize.jpg")
plt.show()
```

#### Using all available features - Logistic Regression

```python
log_model = LogisticRegression()
log_model.fit(train_data[all_available_features], train_data['class'])
log_pred = log_model.predict(test_data[all_available_features])
log_cm = confusion_matrix(test_data['class'], log_pred)
log_disp = ConfusionMatrixDisplay(confusion_matrix=log_cm)
log_disp.plot()
plt.title('Confusion Matrix of Logistic Regression with all Available Features')
plt.savefig("Generated_Graphics/CM_LR_AllFeatures.jpg")
plt.show()
```

#### Using all available features - K Nearest Neighbors

```python
knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(train_data[all_available_features], train_data['class'])
knn_pred = knn_model.predict(test_data[all_available_features])
knn_cm = confusion_matrix(test_data['class'], knn_pred)
knn_disp = ConfusionMatrixDisplay(confusion_matrix=knn_cm)
knn_disp.plot()
plt.title('Confusion Matrix of K Nearest Neighbors with all Available Features')
plt.savefig("Generated_Graphics/CM_KNN_AllFeatures.jpg")
plt.show()
```

#### Using all available features - AdaBoost

```python
ab_model = AdaBoostClassifier(n_estimators = 10)
ab_model.fit(train_data[all_available_features], train_data['class'])
ab_pred = ab_model.predict(test_data[all_available_features])
ab_cm = confusion_matrix(test_data['class'], ab_pred)
ab_disp = ConfusionMatrixDisplay(confusion_matrix=ab_cm)
ab_disp.plot()
plt.title('Confusion Matrix of AdaBoost with all Available Features')
plt.savefig("Generated_Graphics/CM_AB_AllFeatures.jpg")
plt.show()
```

#### Summary of Findings in the Data Modeling part
1. We built a total of six models to determine edibility on two sets of features. The first set includes all the size and color features we explored in the previous part, and we used logistic regression, k nearest neighbors (KNN), and the AdaBoost model since this is a classification task. Among these three models, we can observe that the KNN model performs the best based on the confusion matrix, then the AdaBoost model, and at last logistic regression model. 
2. We built three more models based on all the available features (features that do not contain NA values) with logistic regression, KNN, and AdaBoost, and the performance of these three models is compared to the same model built with only size and color features, which again shows that color and size features are not deterministic regarding the edibility of mushrooms.
3. Among two sets of features, KNN performs well in general, and since the KNN model predicts the results based on how the features of one mushroom are close to other mushrooms, we conclude that mushrooms with similar appearances tend to have the same edibility. A mushroom that looks similar to other edible mushrooms tends to be edible, and a mushroom that looks similar to other poisonous mushrooms tends to be poisonous.
