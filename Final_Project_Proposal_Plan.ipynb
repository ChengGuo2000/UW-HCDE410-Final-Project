### Data
The dataset I used for this project comes from this website: https://archive.ics.uci.edu/ml/datasets/Secondary+Mushroom+Dataset. It is suitable for my project because it listed the information of 61069 mushrooms, for each of them, 17 nominal and 3 metric appearance features and a target (edible or poisonous) are provided, which is perfect for my goal of building a model that can classify a mushroom's edibility based on its appearance. The data is using a Creative Commons license, while the author stated that it is the largest dataset available for determining the edibility of mushrooms. All the information is generated hypothetically from known knowledge of mushrooms for the classification task, so it can be used as a reference for mushrooms while not totally accurate.

## A5 - Final Project Plan

### Overview, Research Questions, and Hypotheses
The goal of this project is about building a machine-learning model that can predict a mushroom's edibility based on its appearance. The reason I want to conduct this project is that I have friends who live in areas where there are abundant mushrooms, and I am eager to learn about how to choose edible mushrooms based on their appearances in the wild, so I want to build this model based on the dataset I found. 

**Research Question:** How does the appearance of a mushroom related to whether it is edible or poisonous?

**Research Hypothesis:** The mushrooms with bigger sizes and brighter colors are more likely to be poisonous than smaller, uncolorful mushrooms, which tend to be edible. 

### Background and Related Work
The dataset I used is generated by a mushroom researcher, Dennis Wagner, in 2021. While it is generated with existing knowledge related to mushrooms, it is inspired by a dataset in 1987, which is also about mushrooms. I found that the 1987 dataset contains fewer data, also there already is enough research done on that one, so I chose this new, bigger dataset for performing data analytics. The research done on the 1987 dataset is mostly related to data analysis skills, including building a Naive Bayes classifier or a neural network, rather than the mushroom itself. My project is focused on the nature of mushrooms and I am building a model that could help people to classify mushrooms in real life. For mushrooms, what I currently know about them is that those which are too colorful must be poisonous, but this statement has not yet been validated with data, so I would like to validate this theory, also use other attributes in this dataset to build a more comprehensive machine-learning model. 

### Methdology
There will be two parts in this project. 
1. The first one is data exploration phrase, where I make some visualizations based on the mushroom appearance data. For this part, I intended to explore the relationship between the color and size of different parts of the mushroom (metrics like cap color, gill color, stem color, veil color, cap diameter, stem height, and stem width) and its edibility. I intended to use percentage stacked bar charts for any qualitative metric and edibility, and a colored scatterplot for quantitative metrics and edibility. The color and size metrics are the one most easy to observe or measure when you encounter a mushroom in the wild, so based on these visualizations, I can get the distribution intuitively and have a basic idea of how to determine a mushroom's edibility based on its appearances.
2. The second part of this project is about data modeling, where I build a machine-learning model to predict a mushroom's edibility based on its appearances. I would build two model, the first one using only the color and size data, and the second one using all 20 attributes in the dataset. Since I am predicting whether a mushroom is edible or not, I would choose from logistic regression, k nearest neighbors, and decision trees to build a model with the higher F-score to predict the edibility. The reason I use F-score is that we want to minimize both false positives and false negatives in our final model, and a higher F-score means both the precision and recall are high.
